## Function

```python
def scrape_traffic_data(road_name):

    file_path = '../data/raw/'
    file_name = road_name + '.traffic.htm'

    # Read the HTML content
    with open(file_path + file_name, 'r', encoding='utf-8') as file:
        html_content = file.read()

    # Create a Beautiful Soup object
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find the table in the soup object
    table = soup.find('table', {'class': 'clsTbl'})

    # Extract table rows
    rows = table.find_all('tr')

    # Extract table data
    table_data = []
    for row in rows:
        cols = row.find_all(['td', 'th'])
        cols = [col.text.strip() for col in cols]
        table_data.append(cols)

    # Assign correct column names
    table_data[4][0] = 'Link no'
    table_data[4][1] = 'Name'
    table_data[4][2] = 'S_LRP'
    table_data[4][3] = 'S_Offset'
    table_data[4][4] = 'S_Chainage'
    table_data[4][5] = 'E_LRP'
    table_data[4][6] = 'E_Offset'
    table_data[4][7] = 'E_Chainage'
    table_data[4][8] = 'Length (km)'

    # Remove the first 3 elements from the list
    table_data = table_data[4:]

    # Remove the last 2 elements from the list
    table_data = table_data[:-2]

    # Create a DataFrame from the table data
    df = pd.DataFrame(table_data[1:], columns=table_data[0])

    return df

# Test the function
N1 = scrape_traffic_data('N1')
```

## Scrape all roads

```python
roads = ['N1', 'N102', 'N104', 'N105', 'N106', 'N2', 'N204', 'N207', 'N208']

for road in roads:
    df = scrape_traffic_data(road)
    print(road, df.shape)

    # Save the DataFrame to a CSV file
    file_path = '../data/processed/'
    file_name = road + '_traffic.csv'
    df.to_csv(file_path + file_name, index=False)
```
