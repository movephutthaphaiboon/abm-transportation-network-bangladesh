# Clean Demo File

To filter only roads longer than 25 km

```python
import pandas as pd
import numpy as np
```

```python
## Import lrp data
file_name = 'df_road_N1andN2.csv'
df_demo = pd.read_csv('../data/' + file_name) #df_road_N1andN2

df_demo
```

```python
selected_roads = ['N1', 'N102', 'N104', 'N105', 'N106', 'N2', 'N204', 'N207', 'N208'] # roads longer than 25 km; knowledge from Assignment 3
df_filtered = df_demo[df_demo['road'].isin(selected_roads)]
df_filtered
```

```python
# Save the filtered DataFrame to a new CSV file
file_path = '../data/processed/'
output_file_name = 'demo_100.csv'
df_filtered.to_csv(file_path + output_file_name, index=False)
```
