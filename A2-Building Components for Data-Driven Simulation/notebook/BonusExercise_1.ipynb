```python
import pandas as pd
import numpy as np
```

```python
# import data
dir = '../data/processed/'
filename = 'demo_100.csv'
df_demo = pd.read_csv(dir + filename)
```

```python
# keep only the model_type with bridge
df_demo_bridge = df_demo.copy()
df_demo_bridge = df_demo_bridge[df_demo_bridge['model_type'] == 'bridge']

# keep only the road with N1
df_demo_bridge = df_demo_bridge[df_demo_bridge['road'] == 'N1']
```

```python
df_demo_bridge.head()
```

```python
df_demo_bridge.sort_values(by='length', ascending=False).head()
```

```python
# Define scoring system
condition_priority = {'A': 1, 'B': 2, 'C': 3, 'D': 4}  # Higher = worse condition

# Normalize bridge length (scale 1-4 based on max length)
df_demo_bridge['normalized_length'] = (df_demo_bridge['length'] / df_demo_bridge['length'].max()) * 4

# Assign condition scores
df_demo_bridge['condition_score'] = df_demo_bridge['condition'].map(condition_priority)

# Define weights (adjust as needed)
W1 = 0.7  # Weight for condition
W2 = 0.3  # Weight for length

# Compute final prioritization score
df_demo_bridge['priority_score'] = (
    W1 * df_demo_bridge['condition_score'] +
    W2 * df_demo_bridge['normalized_length'] 
)

# Rank bridges based on priority score
df_demo_bridge = df_demo_bridge.sort_values(by='priority_score', ascending=False).reset_index(drop=True)

# Select top 5 bridges
top_5_bridges = df_demo_bridge.head(5)
top_5_bridges
```
